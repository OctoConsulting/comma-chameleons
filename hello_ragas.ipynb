```python
def get_context(item):
    titles, ids = item["supporting_facts"].values()
    title_ids = [item["context"]["title"].index(i) for i in titles]
    sentences = [
        item["context"]["sentences"][i][k]
        for i, k in zip(title_ids, item["supporting_facts"]["sent_id"])
    ]
    orig_context = " ".join(sentences)
    return {"answer_context": orig_context}
ds = ds.map(lambda x: get_context(x), batched=False)
```

```python
def get_all_sentences(item):
    all_sentences = [sentence for sublist in item["context"]["sentences"] for sentence in sublist]
    all_context = " ".join(all_sentences)
    return {"all_context": all_context}

ds = ds.map(lambda x: get_all_sentences(x), batched=False)
```
